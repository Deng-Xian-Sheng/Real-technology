### Unsloth

````python
from unsloth import FastLanguageModel  # ä» unsloth åº“å¯¼å…¥ FastLanguageModel ç±»
import torch  # å¯¼å…¥ PyTorch åº“

max_seq_length = 2048  # è®¾ç½®æœ€å¤§åºåˆ—é•¿åº¦ï¼Œæ­¤å¤„ä¸º 2048ï¼Œæ”¯æŒ RoPE Scaling è‡ªåŠ¨ç¼©æ”¾
dtype = None  # æ•°æ®ç±»å‹ï¼Œè®¾ç½®ä¸º None è‡ªåŠ¨æ£€æµ‹ï¼ŒTesla T4ã€V100 ä½¿ç”¨ Float16ï¼ŒAmpere+ ä½¿ç”¨ Bfloat16
load_in_4bit = True  # ä½¿ç”¨ 4 ä½é‡åŒ–å‡å°‘å†…å­˜ä½¿ç”¨ï¼Œå¯ä»¥è®¾ç½®ä¸º False

# æ”¯æŒçš„ 4 ä½é¢„é‡åŒ–æ¨¡å‹ï¼Œä¸‹è½½é€Ÿåº¦æå‡ 4 å€ï¼Œé¿å…å†…å­˜ä¸è¶³é”™è¯¯
fourbit_models = [
    "unsloth/Meta-Llama-3.1-8B-bnb-4bit",  # Llama-3.1ï¼Œé€Ÿåº¦æå‡ 2 å€
    "unsloth/Meta-Llama-3.1-70B-bnb-4bit",
    "unsloth/Mistral-Small-Instruct-2409",  # Mistral 22bï¼Œé€Ÿåº¦æå‡ 2 å€ï¼
    "unsloth/mistral-7b-instruct-v0.3-bnb-4bit",
    "unsloth/Phi-3.5-mini-instruct",  # Phi-3.5ï¼Œé€Ÿåº¦æå‡ 2 å€ï¼
    "unsloth/Phi-3-medium-4k-instruct",
    "unsloth/gemma-2-27b-bnb-4bit",  # Gemmaï¼Œé€Ÿåº¦æå‡ 2 å€ï¼
    "unsloth/Llama-3.2-1B-bnb-4bit",  # å…¨æ–°ï¼Llama 3.2 æ¨¡å‹
    "unsloth/Llama-3.2-1B-Instruct-bnb-4bit",
    "unsloth/Llama-3.2-3B-Instruct-bnb-4bit",
]  # æ›´å¤šæ¨¡å‹è¯·è®¿é—® https://huggingface.co/unsloth

qwen_models = [
    "unsloth/Qwen2.5-Coder-32B-Instruct",  # Qwen 2.5 Coderï¼Œé€Ÿåº¦æå‡ 2 å€
    "unsloth/Qwen2.5-Coder-7B",
    "unsloth/Qwen2.5-14B-Instruct",  # 14B å‚æ•°æ¨¡å‹é€‚ç”¨äº 16GB æ˜¾å¡
    "unsloth/Qwen2.5-7B",
    "unsloth/Qwen2.5-72B-Instruct",  # 72B å‚æ•°æ¨¡å‹é€‚ç”¨äº 48GB æ˜¾å¡
    "unsloth/DeepSeek-R1-Distill-Qwen-1.5B-unsloth-bnb-4bit"
]  # æ›´å¤šæ¨¡å‹è¯·è®¿é—® https://huggingface.co/unsloth

# ä»é¢„è®­ç»ƒæ¨¡å‹åŠ è½½æ¨¡å‹å’Œåˆ†è¯å™¨
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/DeepSeek-R1-Distill-Qwen-1.5B-unsloth-bnb-4bit",  # æ¨¡å‹åç§°
    max_seq_length=max_seq_length,  # æœ€å¤§åºåˆ—é•¿åº¦
    dtype=dtype,  # æ•°æ®ç±»å‹
    load_in_4bit=load_in_4bit,  # æ˜¯å¦ä½¿ç”¨ 4 ä½é‡åŒ–åŠ è½½
    # token="hf_...",  # å¦‚æœä½¿ç”¨ gated æ¨¡å‹ï¼ˆå¦‚ meta-llama/Llama-2-7b-hfï¼‰ï¼Œè¯·ä½¿ç”¨æ­¤å‚æ•°
)
````

æˆ‘ä»¬ç°åœ¨æ·»åŠ  LoRA é€‚é…å™¨ï¼Œè¿™æ ·æˆ‘ä»¬åªéœ€è¦æ›´æ–° 1% åˆ° 10% çš„æ‰€æœ‰å‚æ•°ï¼

````python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,  # é€‰æ‹©ä»»æ„å¤§äº 0 çš„æ•°å­—ï¼å»ºè®®å€¼ï¼š8ã€16ã€32ã€64ã€128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],  # åº”ç”¨ LoRA çš„ç›®æ ‡æ¨¡å—
    lora_alpha = 16,  # LoRA çš„ç¼©æ”¾å› å­
    lora_dropout = 0,  # æ”¯æŒä»»æ„å€¼ï¼Œä½† 0 æ˜¯æœ€ä¼˜åŒ–çš„
    bias = "none",  # æ”¯æŒä»»æ„å€¼ï¼Œä½† "none" æ˜¯æœ€ä¼˜åŒ–çš„
    # [æ–°åŠŸèƒ½] "unsloth" å‡å°‘ 30% çš„æ˜¾å­˜ä½¿ç”¨ï¼Œé€‚ç”¨äº 2 å€å¤§çš„æ‰¹æ¬¡å¤§å°ï¼
    use_gradient_checkpointing = "unsloth",  # å¯¹äºéå¸¸é•¿çš„ä¸Šä¸‹æ–‡ï¼Œè®¾ç½®ä¸º True æˆ– "unsloth"
    random_state = 3407,  # éšæœºç§å­
    use_rslora = False,  # æ”¯æŒç§©ç¨³å®š LoRA
    loftq_config = None,  # æ”¯æŒ LoftQ
)
````

## æ•°æ®å‡†å¤‡
ç°åœ¨ï¼Œæˆ‘ä»¬ä½¿ç”¨ Qwen-2.5 æ ¼å¼è¿›è¡Œå¯¹è¯é£æ ¼çš„å¾®è°ƒã€‚æˆ‘ä»¬ä½¿ç”¨ [redacted link] æ•°æ®é›†ï¼Œè¯¥æ•°æ®é›†é‡‡ç”¨ ShareGPT é£æ ¼ã€‚ä½†æ˜¯ï¼Œæˆ‘ä»¬å°†å…¶è½¬æ¢ä¸º HuggingFace çš„æ ‡å‡†å¤šè½®å¯¹è¯æ ¼å¼ ("role", "content")ï¼Œè€Œä¸æ˜¯ ("from", "value")ã€‚Qwen æ¸²æŸ“å¤šè½®å¯¹è¯çš„æ–¹å¼å¦‚ä¸‹ï¼š

```
<|im_start|>system
ä½ æ˜¯æ–‡å¿ƒä¸€è¨€ï¼Œç”±é˜¿é‡Œäº‘åˆ›é€ ã€‚ä½ æ˜¯ä¸€ä¸ªä¹äºåŠ©äººçš„åŠ©æ‰‹ã€‚<|im_end|>
<|im_start|>user
2+2 ç­‰äºå¤šå°‘ï¼Ÿ<|im_end|>
<|im_start|>assistant
ç­‰äº 4ã€‚<|im_end|>
```

æˆ‘ä»¬ä½¿ç”¨ get_chat_template å‡½æ•°æ¥è·å–æ­£ç¡®çš„å¯¹è¯æ¨¡æ¿ã€‚æˆ‘ä»¬æ”¯æŒ zephyrã€chatmlã€mistralã€llamaã€alpacaã€vicunaã€vicuna_oldã€phi3ã€llama3 ç­‰å¤šç§å¯¹è¯æ¨¡æ¿ã€‚

````python
from unsloth.chat_templates import get_chat_template

tokenizer = get_chat_template(
    tokenizer,
    chat_template = "qwen-2.5",
)

def formatting_prompts_func(examples):
    convos = examples["conversations"]
    texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False) for convo in convos]
    return { "text" : texts, }
pass

from datasets import load_dataset
dataset = load_dataset("mlabonne/FineTome-100k", split = "train")
````

æˆ‘ä»¬ç°åœ¨ä½¿ç”¨ standardize_sharegpt å°† ShareGPT é£æ ¼çš„æ•°æ®é›†è½¬æ¢ä¸º HuggingFace çš„é€šç”¨æ ¼å¼ã€‚è¿™ä¼šå°†æ•°æ®é›†ä»ä»¥ä¸‹å½¢å¼ï¼š

```
{"from": "system", "value": "You are an assistant"}
{"from": "human", "value": "What is 2+2?"}
{"from": "gpt", "value": "It's 4."}
```
è½¬æ¢ä¸ºä»¥ä¸‹å½¢å¼ï¼š
```
{"role": "system", "content": "You are an assistant"}
{"role": "user", "content": "What is 2+2?"}
{"role": "assistant", "content": "It's 4."}
```


````python
from unsloth.chat_templates import standardize_sharegpt
dataset = standardize_sharegpt(dataset)
dataset = dataset.map(formatting_prompts_func, batched = True,)
````

# æ¨¡å‹è®­ç»ƒ
ç°åœ¨ï¼Œè®©æˆ‘ä»¬ä½¿ç”¨ Huggingface TRL åº“çš„ SFTTrainerï¼æ›´å¤šæ–‡æ¡£è¯·å‚è€ƒï¼š[redacted link]ã€‚ä¸ºäº†åŠ å¿«è®­ç»ƒé€Ÿåº¦ï¼Œæˆ‘ä»¬åªè¿›è¡Œ 60 æ­¥è¿­ä»£ï¼Œä½†æ‚¨å¯ä»¥è®¾ç½® num_train_epochs=1 è¿›è¡Œå®Œæ•´è®­ç»ƒï¼Œå¹¶å…³é—­ max_steps=Noneã€‚æˆ‘ä»¬ä¹Ÿæ”¯æŒ TRL çš„ DPOTrainerï¼

æ­¤è®­ç»ƒå™¨åŒ…å«æˆ‘ä»¬å¯¹ æ¢¯åº¦ç´¯ç§¯é”™è¯¯ çš„ä¿®å¤ã€‚æƒ³äº†è§£æ›´å¤šä¿¡æ¯ï¼Œè¯·é˜…è¯»ï¼š[redacted link]

````python
from trl import SFTTrainer  # ä» trl åº“å¯¼å…¥ SFTTrainer ç±»ï¼Œç”¨äºç›‘ç£å¼å¾®è°ƒ
from transformers import TrainingArguments, DataCollatorForSeq2Seq  # ä» transformers åº“å¯¼å…¥è®­ç»ƒå‚æ•°å’Œæ•°æ®æ•´ç†å™¨
from unsloth import is_bfloat16_supported  # ä» unsloth åº“å¯¼å…¥ is_bfloat16_supported å‡½æ•°ï¼Œç”¨äºæ£€æŸ¥æ˜¯å¦æ”¯æŒ bf16 æ•°æ®ç±»å‹


trainer = SFTTrainer(  # åˆ›å»ºä¸€ä¸ª SFTTrainer å®ä¾‹
    model=model,  # è¦å¾®è°ƒçš„æ¨¡å‹
    tokenizer=tokenizer,  # åˆ†è¯å™¨
    train_dataset=dataset,  # è®­ç»ƒæ•°æ®é›†
    dataset_text_field="text",  # æ•°æ®é›†ä¸­æ–‡æœ¬å­—æ®µçš„åç§°
    max_seq_length=max_seq_length,  # æœ€å¤§åºåˆ—é•¿åº¦
    data_collator=DataCollatorForSeq2Seq(tokenizer=tokenizer),  # æ•°æ®æ•´ç†å™¨ï¼Œç”¨äºå°†æ•°æ®æ•´ç†æˆæ¨¡å‹æ‰€éœ€çš„æ ¼å¼
    dataset_num_proc=4,  # ç”¨äºæ•°æ®å¤„ç†çš„è¿›ç¨‹æ•°
    packing=False,  # æ˜¯å¦æ‰“åŒ…æ•°æ®ï¼Œå¯¹äºçŸ­åºåˆ—å¯ä»¥åŠ å¿«è®­ç»ƒé€Ÿåº¦ 5 å€
    args=TrainingArguments(  # è®­ç»ƒå‚æ•°
        per_device_train_batch_size=1,  # æ¯ä¸ªè®¾å¤‡çš„è®­ç»ƒæ‰¹æ¬¡å¤§å°
        gradient_accumulation_steps=4,  # æ¢¯åº¦ç´¯ç§¯æ­¥æ•°ï¼Œæœ€æ–°ç‰ˆ Unsloth ä¿®å¤äº†ä¸€ä¸ªä¸»è¦é”™è¯¯
        warmup_steps=5,  # é¢„çƒ­æ­¥æ•°
        # num_train_epochs=1,  # è®­ç»ƒè½®æ•°ï¼Œè®¾ç½®ä¸º 1 è¿›è¡Œå®Œæ•´è®­ç»ƒ
        max_steps=30,  # æœ€å¤§è®­ç»ƒæ­¥æ•°
        learning_rate=2e-4,  # å­¦ä¹ ç‡
        fp16=not is_bfloat16_supported(),  # æ˜¯å¦ä½¿ç”¨ fp16 æ•°æ®ç±»å‹
        bf16=is_bfloat16_supported(),  # æ˜¯å¦ä½¿ç”¨ bf16 æ•°æ®ç±»å‹
        logging_steps=1,  # æ—¥å¿—è®°å½•æ­¥æ•°
        optim="paged_adamw_8bit",  # ä¼˜åŒ–å™¨ï¼Œä½¿ç”¨ paged_adamw_8bit èŠ‚çœæ›´å¤šå†…å­˜
        weight_decay=0.01,  # æƒé‡è¡°å‡
        lr_scheduler_type="linear",  # å­¦ä¹ ç‡è°ƒåº¦å™¨ç±»å‹
        seed=3407,  # éšæœºç§å­
        output_dir="outputs",  # è¾“å‡ºç›®å½•
        report_to="none",  # æ˜¯å¦å°†è®­ç»ƒæŠ¥å‘Šå‘é€åˆ°å…¶ä»–å¹³å°ï¼Œä¾‹å¦‚ WandB
    ),
)
````

æˆ‘ä»¬è¿˜ä½¿ç”¨ Unsloth çš„ train_on_completions æ–¹æ³•ï¼Œä»…å¯¹åŠ©æ‰‹çš„è¾“å‡ºè¿›è¡Œè®­ç»ƒï¼Œè€Œå¿½ç•¥ç”¨æˆ·è¾“å…¥çš„æŸå¤±ã€‚

````python
from unsloth.chat_templates import train_on_responses_only
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<|im_start|>user\n",
    response_part = "<|im_start|>assistant\n",
)
````

## æ¨ç†
è®©æˆ‘ä»¬è¿è¡Œæ¨¡å‹ï¼æ‚¨å¯ä»¥æ›´æ”¹æŒ‡ä»¤å’Œè¾“å…¥ - å°†è¾“å‡ºç•™ç©ºï¼

[æ–°åŠŸèƒ½] åœ¨å…è´¹çš„ Colab ä¸­å°è¯•ä½¿ç”¨ Llama-3.1 8b Instruct è¿›è¡Œ 2 å€é€Ÿæ¨ç† [æ­¤å¤„](å·²åˆ é™¤é“¾æ¥)

æˆ‘ä»¬ä½¿ç”¨ min_p = 0.1 å’Œ temperature = 1.5ã€‚é˜…è¯»æ­¤ [æ¨æ–‡] (å·²åˆ é™¤é“¾æ¥)ä»¥äº†è§£æ›´å¤šä¿¡æ¯ã€‚

````python
from unsloth.chat_templates import get_chat_template  # ä» unsloth åº“å¯¼å…¥ get_chat_template å‡½æ•°

# è·å– Qwen-2.5 çš„èŠå¤©æ¨¡æ¿ï¼Œå¹¶åº”ç”¨åˆ°åˆ†è¯å™¨
tokenizer = get_chat_template(
    tokenizer,
    chat_template="qwen-2.5",
)

FastLanguageModel.for_inference(model)  # å¯ç”¨åŸç”Ÿ 2 å€é€Ÿæ¨ç†

# å®šä¹‰æ¶ˆæ¯åˆ—è¡¨ï¼ŒåŒ…å«ç”¨æˆ·è§’è‰²å’Œå†…å®¹
messages = [
    {"role": "user", "content": "Continue the fibonnaci sequence: 1, 1, 2, 3, 5, 8,"},  # ç”¨æˆ·è¦æ±‚ç»§ç»­æ–æ³¢é‚£å¥‘æ•°åˆ—
]

# ä½¿ç”¨èŠå¤©æ¨¡æ¿å¤„ç†æ¶ˆæ¯ï¼Œè¿›è¡Œåˆ†è¯ï¼Œæ·»åŠ ç”Ÿæˆæç¤ºï¼Œå¹¶è½¬æ¢ä¸º PyTorch å¼ é‡
inputs = tokenizer.apply_chat_template(
    messages,
    tokenize=True,
    add_generation_prompt=True,  # å¿…é¡»æ·»åŠ ç”¨äºç”Ÿæˆ
    return_tensors="pt",
).to("cuda")  # å°†å¼ é‡ç§»åŠ¨åˆ° CUDA è®¾å¤‡ï¼ˆGPUï¼‰

# ä½¿ç”¨æ¨¡å‹ç”Ÿæˆè¾“å‡º
outputs = model.generate(
    input_ids=inputs,  # è¾“å…¥ ID
    max_new_tokens=64,  # æœ€å¤§ç”Ÿæˆçš„æ–°æ ‡è®°æ•°
    use_cache=True,  # ä½¿ç”¨ç¼“å­˜åŠ é€Ÿç”Ÿæˆ
    temperature=1.5,  # æ§åˆ¶ç”Ÿæˆç»“æœçš„éšæœºæ€§
    min_p=0.1,  # æ§åˆ¶ç”Ÿæˆç»“æœçš„å¤šæ ·æ€§
)

# å°†æ¨¡å‹è¾“å‡ºè§£ç ä¸ºæ–‡æœ¬
tokenizer.batch_decode(outputs)
````

æ‚¨ä¹Ÿå¯ä»¥ä½¿ç”¨ TextStreamer è¿›è¡Œè¿ç»­æ¨ç† - è¿™æ ·æ‚¨å°±å¯ä»¥é€ä¸ªæ ‡è®°åœ°çœ‹åˆ°ç”Ÿæˆç»“æœï¼Œè€Œä¸æ˜¯ä¸€ç›´ç­‰å¾…ï¼

````python
FastLanguageModel.for_inference(model) # å¯ç”¨åŸç”Ÿ 2 å€é€Ÿæ¨ç†

messages = [
    {"role": "user", "content": "Continue the fibonnaci sequence: 1, 1, 2, 3, 5, 8,"},  # ç”¨æˆ·æ¶ˆæ¯ï¼šç»§ç»­æ–æ³¢é‚£å¥‘æ•°åˆ—
]
inputs = tokenizer.apply_chat_template(
    messages,
    tokenize = True,  # å¯¹æ¶ˆæ¯è¿›è¡Œåˆ†è¯
    add_generation_prompt = True,  # å¿…é¡»æ·»åŠ ç”¨äºç”Ÿæˆ
    return_tensors = "pt",  # è¿”å› PyTorch å¼ é‡
).to("cuda")  # å°†å¼ é‡ç§»åŠ¨åˆ° CUDA è®¾å¤‡ï¼ˆGPUï¼‰

from transformers import TextStreamer  # å¯¼å…¥ TextStreamer ç”¨äºè¿ç»­æ¨ç†
text_streamer = TextStreamer(tokenizer, skip_prompt = True)  # åˆ›å»º TextStreamer å®ä¾‹
_ = model.generate(input_ids = inputs, streamer = text_streamer, max_new_tokens = 128,  # ä½¿ç”¨æ¨¡å‹ç”Ÿæˆè¾“å‡º
                   use_cache = True, temperature = 1.5, min_p = 0.1)  # è®¾ç½®ç”Ÿæˆå‚æ•°
````

### ä¿å­˜å’ŒåŠ è½½å¾®è°ƒæ¨¡å‹
è¦å°†æœ€ç»ˆæ¨¡å‹ä¿å­˜ä¸º LoRA é€‚é…å™¨ï¼Œå¯ä»¥ä½¿ç”¨ Huggingface çš„ push_to_hub è¿›è¡Œåœ¨çº¿ä¿å­˜ï¼Œæˆ–ä½¿ç”¨ save_pretrained è¿›è¡Œæœ¬åœ°ä¿å­˜ã€‚

[æ³¨æ„] è¿™åªä¼šä¿å­˜ LoRA é€‚é…å™¨ï¼Œè€Œä¸æ˜¯å®Œæ•´çš„æ¨¡å‹ã€‚è¦ä¿å­˜ä¸º 16 ä½æˆ– GGUF æ ¼å¼ï¼Œè¯·å‘ä¸‹æ»šåŠ¨ï¼

````python
model.save_pretrained("lora_model")  # æœ¬åœ°ä¿å­˜
tokenizer.save_pretrained("lora_model")
# model.push_to_hub("your_name/lora_model", token = "...") # åœ¨çº¿ä¿å­˜
# tokenizer.push_to_hub("your_name/lora_model", token = "...") # åœ¨çº¿ä¿å­˜
````

ç°åœ¨ï¼Œå¦‚æœæ‚¨æƒ³åŠ è½½æˆ‘ä»¬åˆšåˆšä¿å­˜çš„ LoRA é€‚é…å™¨è¿›è¡Œæ¨ç†ï¼Œè¯·å°† False è®¾ç½®ä¸º Trueï¼š

````python
if False:  # è¿™æ®µä»£ç ç›®å‰ä¸ä¼šè¢«æ‰§è¡Œï¼Œå› ä¸ºæ¡ä»¶ä¸º False
    from unsloth import FastLanguageModel  # ä» unsloth åº“å¯¼å…¥ FastLanguageModel ç±»
    # ä»é¢„è®­ç»ƒæ¨¡å‹åŠ è½½æ¨¡å‹å’Œåˆ†è¯å™¨
    model, tokenizer = FastLanguageModel.from_pretrained(
        model_name="lora_model",  # ä½ ç”¨äºè®­ç»ƒçš„æ¨¡å‹åç§°ï¼Œåº”æ›¿æ¢ä¸ºå®é™…çš„æ¨¡å‹åç§°
        max_seq_length=max_seq_length,  # æœ€å¤§åºåˆ—é•¿åº¦ï¼Œä¹‹å‰å·²å®šä¹‰
        dtype=dtype,  # æ•°æ®ç±»å‹ï¼Œä¹‹å‰å·²å®šä¹‰
        load_in_4bit=load_in_4bit,  # æ˜¯å¦ä½¿ç”¨ 4 ä½é‡åŒ–åŠ è½½ï¼Œä¹‹å‰å·²å®šä¹‰
    )
    FastLanguageModel.for_inference(model)  # å¯ç”¨åŸç”Ÿ 2 å€é€Ÿæ¨ç†

# å®šä¹‰æ¶ˆæ¯åˆ—è¡¨ï¼ŒåŒ…å«ç”¨æˆ·è§’è‰²å’Œå†…å®¹
messages = [
    {"role": "user", "content": "Describe a tall tower in the capital of France."},  # ç”¨æˆ·æ¶ˆæ¯ï¼šæè¿°æ³•å›½é¦–éƒ½çš„ä¸€åº§é«˜å¡”
]

# ä½¿ç”¨èŠå¤©æ¨¡æ¿å¤„ç†æ¶ˆæ¯ï¼Œè¿›è¡Œåˆ†è¯ï¼Œæ·»åŠ ç”Ÿæˆæç¤ºï¼Œå¹¶è½¬æ¢ä¸º PyTorch å¼ é‡
inputs = tokenizer.apply_chat_template(
    messages,
    tokenize=True,  # å¯¹æ¶ˆæ¯è¿›è¡Œåˆ†è¯
    add_generation_prompt=True,  # å¿…é¡»æ·»åŠ ç”¨äºç”Ÿæˆ
    return_tensors="pt",  # è¿”å› PyTorch å¼ é‡
).to("cuda")  # å°†å¼ é‡ç§»åŠ¨åˆ° CUDA è®¾å¤‡ï¼ˆGPUï¼‰

from transformers import TextStreamer  # å¯¼å…¥ TextStreamer ç”¨äºè¿ç»­æ¨ç†
text_streamer = TextStreamer(tokenizer, skip_prompt=True)  # åˆ›å»º TextStreamer å®ä¾‹
# ä½¿ç”¨æ¨¡å‹ç”Ÿæˆè¾“å‡º
_ = model.generate(
    input_ids=inputs,  # è¾“å…¥ ID
    streamer=text_streamer,  # ä½¿ç”¨ TextStreamer è¿›è¡Œè¿ç»­æ¨ç†
    max_new_tokens=128,  # æœ€å¤§ç”Ÿæˆçš„æ–°æ ‡è®°æ•°
    use_cache=True,  # ä½¿ç”¨ç¼“å­˜åŠ é€Ÿç”Ÿæˆ
    temperature=1.5,  # æ§åˆ¶ç”Ÿæˆç»“æœçš„éšæœºæ€§
    min_p=0.1,  # æ§åˆ¶ç”Ÿæˆç»“æœçš„å¤šæ ·æ€§
)
````

æ‚¨ä¹Ÿå¯ä»¥ä½¿ç”¨ Hugging Face çš„ AutoModelForPeftCausalLMã€‚ä»…å½“æ‚¨æœªå®‰è£… unsloth æ—¶æ‰ä½¿ç”¨æ­¤æ–¹æ³•ã€‚å®ƒå¯èƒ½ä¼šéå¸¸æ…¢ï¼Œå› ä¸ºå®ƒä¸æ”¯æŒ 4bit æ¨¡å‹ä¸‹è½½ï¼Œè€Œä¸” Unsloth çš„ æ¨ç†é€Ÿåº¦å¿« 2 å€ã€‚



````python
if False:  # è¿™æ®µä»£ç ç›®å‰ä¸ä¼šè¢«æ‰§è¡Œï¼Œå› ä¸ºæ¡ä»¶ä¸º False
    # æˆ‘å¼ºçƒˆä¸å»ºè®®è¿™æ ·åš - å¦‚æœå¯èƒ½ï¼Œè¯·ä½¿ç”¨ Unsloth
    from peft import AutoPeftModelForCausalLM  # ä» peft åº“å¯¼å…¥ AutoPeftModelForCausalLM ç±»
    from transformers import AutoTokenizer  # ä» transformers åº“å¯¼å…¥ AutoTokenizer ç±»

    # ä»é¢„è®­ç»ƒæ¨¡å‹åŠ è½½æ¨¡å‹ï¼Œä½¿ç”¨ LoRA é€‚é…å™¨
    model = AutoPeftModelForCausalLM.from_pretrained(
        "lora_model",  # ä½ ç”¨äºè®­ç»ƒçš„æ¨¡å‹åç§°ï¼Œåº”æ›¿æ¢ä¸ºå®é™…çš„æ¨¡å‹åç§°
        load_in_4bit=load_in_4bit,  # æ˜¯å¦ä½¿ç”¨ 4 ä½é‡åŒ–åŠ è½½ï¼Œä¹‹å‰å·²å®šä¹‰
    )
    # ä»é¢„è®­ç»ƒæ¨¡å‹åŠ è½½åˆ†è¯å™¨
    tokenizer = AutoTokenizer.from_pretrained("lora_model")  # ä½ ç”¨äºè®­ç»ƒçš„æ¨¡å‹åç§°ï¼Œåº”æ›¿æ¢ä¸ºå®é™…çš„æ¨¡å‹
````
